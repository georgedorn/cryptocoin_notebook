```python
%matplotlib inline

from btc_data import data
import matplotlib.pyplot as plot
import numpy as np
import statsmodels.api as sm

first_ts = data[0][0]
seconds_in_day = 60 * 60 * 24

days = []
prices = []
for row in data:
    ts = row[0] - first_ts
    days.append(ts / seconds_in_day)
    prices.append(row[7])
    
#do the regression
y = np.matrix(prices).transpose()
x = np.matrix(days).transpose()
X = sm.add_constant(x)
model = sm.OLS(y, X)
f = model.fit()

print f.summary()

slope = f.params[1]
intercept = f.params[0]
linear = [intercept + slope*day for day in days]

print "Resulting linear regression: price = %s * day + %s" % (slope, intercept)

plot.plot(days, prices)
plot.plot(days, linear)
plot.title("Linear Regression")




;




```

```python
#Now we try a linear regression against the natural log of prices,
#which is the equivalent of an exponential regression

import math
log_prices = [math.log(p) for p in prices]

ln_y = np.matrix(log_prices).transpose()
ln_X = sm.add_constant(x)
ln_model = sm.OLS(ln_y, ln_X)

ln_results = ln_model.fit()
print ln_results.summary()

ln_slope = ln_results.params[1]
ln_intercept = ln_results.params[0]
ln_linear = [ln_intercept + ln_slope*day for day in days]

plot.plot(days, log_prices)
plot.plot(days, ln_linear)
plot.title("Log regression")
;
```
