# Predicting BTC prices via regression modeling.

```python
%matplotlib inline

#Imports
import matplotlib.pyplot as plot
import numpy as np
import math
import statsmodels.api as sm
from datetime import datetime
from copy import copy
from IPython.display import display, Math

#BTC price data, extracted from:
# http://bitcoincharts.com/charts/mtgoxUSD#tgMzbgBza1gWMAzm1g10zm2g25zv
from btc_data import data

#set up data
first_ts = data[0][0] #the first timestamp, everything will be based on this
first_day = datetime.fromtimestamp(first_ts)
print "First day in data set: ", first_day
seconds_in_day = 60 * 60 * 24 #we're only using the weighted average for per-day prices, so we can convert timestamps to days

days = []
prices = []
for row in data:
    ts = row[0] - first_ts
    days.append(ts / seconds_in_day)
    prices.append(row[7])
```

## Attempt 1:  Linear regression

This first regression is just to exercise some libraries, figure out what statsmodels can do.  It's not particularly accurate, as linear regressions are only useful if you expect growth to be constant, like BTC going up 9 cents per day.

```python
#do the regression
y = np.matrix(prices).transpose()
x = np.matrix(days).transpose()
X = sm.add_constant(x)
model = sm.OLS(y, X)
f = model.fit()

slope = f.params[1]
intercept = f.params[0]
linear = [intercept + slope*day for day in days]
equation = Math(r'P(x) = %s x %s%s' % (slope, "+" if intercept>0 else "", intercept))
print "Resulting equation, where x is days since %s:" % first_day.date()
display(equation)
print "R-squared: %s, %s (closer to 1.0 means more accurate)" % (f.rsquared, f.rsquared_adj)

plot.plot(days, prices)
plot.plot(days, linear)
plot.title("Linear Regression")
;
```

So that's not great, obviously.  The R-squared is a pretty good indicator of inaccuracy.

## Attempt 2: Exponential Regression

This regression is actually a linear regression performed on the natural logs of the prices.  The resulting coefficients fit into a formula that looks like this:
$$ P(x) = e^{mx+b} $$

```python
log_prices = [math.log(p) for p in prices]

ln_y = np.matrix(log_prices).transpose()
ln_X = sm.add_constant(x)
ln_model = sm.OLS(ln_y, ln_X)

ln_results = ln_model.fit()

ln_slope = ln_results.params[1]
ln_intercept = ln_results.params[0]
ln_linear = [ln_intercept + ln_slope*day for day in days]

plot.plot(days, log_prices)
plot.plot(days, ln_linear)
plot.title("Linear regression of log(price)")

print "Resulting equation:"
ln_equation = Math(r'P(x) = e^{%sx %s%s}' % (ln_slope, '+' if ln_intercept > 0 else '', ln_intercept))
display(ln_equation)
print "R-squared: %s, %s (closer to 1.0 means more accurate)" % (ln_results.rsquared, ln_results.rsquared_adj)
;
```

That's looking a lot better.  It follows the overall trend, with bubbles being responsible for most deviations.
Let's graph the original prices against our new equation.

```python
last_day = days[-1]

extended_days = copy(days)
extended_prices = copy(prices)
extended_days.extend( [last_day+i for i in range(1,100)] )
extended_prices.extend( [None for i in range(1,100)] )

plot.plot(extended_days, extended_prices)

predict_price = lambda x: math.exp(x*ln_slope + ln_intercept)

predicted_prices = [predict_price(day) for day in extended_days]
plot.plot(extended_days, predicted_prices)
plot.title("BTC price vs exponential model")
;
```
